```python
from sklearn import preprocessing
```

```python
import pandas as pd
```

### scale numeric columns in train data

```python
raw_train = pd.read_csv('./data/train.csv')
```

```python
raw_train.describe()
```

```python
col_num_name = map(lambda x:'cont'+str(x+1),range(14))
col_num_name.insert(0,'loss')
train_scale_num = preprocessing.scale(raw_train.loc[:,col_num_name])
```

```python
train_scale_num = pd.DataFrame(train_scale_num)
train_scale_num.columns = col_num_name
train_scale_num.describe()
```

```python
x = raw_train[raw_train.columns[map(lambda i:i not in col_num_name,raw_train.columns)]]
x.head()
```

```python
train_scale = pd.concat([train_scale_num.reset_index(drop=True),
                         x.reset_index(drop=True)],axis=1)
train_scale.head()
```

```python
train_scale.to_csv('./output/raw_train_scale.csv')
```

### scale numeric data in test data

```python
raw_test = pd.read_csv('./data/test.csv')
```

```python
col_num_name = map(lambda x:'cont'+str(x+1),range(14))
test_scale_num = preprocessing.scale(raw_test.loc[:,col_num_name])
```

```python
test_scale_num = pd.DataFrame(test_scale_num)
test_scale_num.columns = col_num_name
test_scale_num.describe()
```

```python
x = raw_test[raw_test.columns[map(lambda i:i not in col_num_name,
                                  raw_test.columns)]]
test_scale = pd.concat([test_scale_num.reset_index(drop=True),
                         x.reset_index(drop=True)],axis=1)
test_scale.head()
```

```python
test_scale.to_csv('./output/raw_test_scale.csv')
```
